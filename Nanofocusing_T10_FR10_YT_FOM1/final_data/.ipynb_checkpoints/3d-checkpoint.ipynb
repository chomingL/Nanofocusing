```python
import meep as mp
import meep.adjoint as mpa
from meep import Animate2D
import numpy as np
from autograd import numpy as npa
from autograd import tensor_jacobian_product, grad
import nlopt
from matplotlib import pyplot as plt
from matplotlib.patches import Circle
from meep.materials import Ag
import os
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from PIL import Image
import matplotlib.colors as mcolors
from scipy.integrate import trapezoid
from matplotlib.patches import Rectangle
mp.verbosity(1)
TiO2 = mp.Medium(index=2.6)
SiO2 = mp.Medium(index=1.44)
Si = mp.Medium(index=3.4)
Air = mp.Medium(index=1)


resolution = 100
design_region_resolution = int(resolution)

design_region_x_width  = 1   #100 nm
design_region_y_height = 1   #100 nm
design_region_z_height = 0.01  #20 nm or 10nm

pml_size = 1.0
pml_layers = [mp.PML(pml_size,direction=mp.Z)]

Sz_size = 0.6
Sx = design_region_x_width
Sy = design_region_y_height 
Sz = 2 * pml_size + design_region_z_height + Sz_size
cell_size = mp.Vector3(Sx, Sy, Sz)

wavelengths = np.array([1.55])     # wavelengths = np.array([1.5 ,1.55, 1.6])
frequencies = np.array([1 / 1.55])

nf = 1                 #3 #wavelengths Number

minimum_length = 0.01  # minimum length scale (microns)
eta_i = 0.5            # blueprint (or intermediate) design field thresholding point (between 0 and 1)
eta_e = 0.55           # erosion design field thresholding point (between 0 and 1)
eta_d = 1 - eta_e      # dilation design field thresholding point (between 0 and 1)
filter_radius = mpa.get_conic_radius_from_eta_e(minimum_length, eta_e)

Source_distance = -0.2

fcen   = 1 / 1.55
width  = 0.2  
fwidth = width * fcen
source_center = mp.Vector3(0,0,Source_distance)  
source_size   = mp.Vector3(design_region_x_width, design_region_y_height, 0)
src    = mp.GaussianSource(frequency=fcen, fwidth=fwidth, is_integrated=True)
source = [mp.Source(src, component=mp.Ey, size=source_size, center=source_center)]   

Nx = int(design_region_resolution * design_region_x_width) + 1
Ny = int(design_region_resolution * design_region_y_height) + 1
Nz = 1

design_variables = mp.MaterialGrid(mp.Vector3(Nx, Ny, Nz), Air, Ag, grid_type="U_MEAN")
design_region    = mpa.DesignRegion(
            design_variables,
            volume=mp.Volume(
            center=mp.Vector3(0,0,0),
            size=mp.Vector3(design_region_x_width, design_region_y_height, design_region_z_height),
            ),
)

def mapping(x, eta, beta):
    # filter
    filtered_field = mpa.conic_filter(
        x,
        filter_radius,
        design_region_x_width,
        design_region_y_height,
        design_region_resolution,
    )
    # projection
    projected_field = mpa.tanh_projection(filtered_field, beta, eta)
    projected_field = (npa.fliplr(projected_field) + projected_field) / 2
    projected_field = (npa.flipud(projected_field) + projected_field) / 2  # left-right symmetry    
    
    return projected_field.flatten()


geometry = [mp.Block(center=design_region.center, size=design_region.size, material=design_variables)]

kpoint = mp.Vector3()
sim = mp.Simulation(
    cell_size        = cell_size,
    boundary_layers  = pml_layers,
    geometry         = geometry,
    sources          = source,
    default_material = Air,
    k_point          = kpoint,
    symmetries       = [mp.Mirror(direction=mp.X)],
    resolution       = resolution,
    extra_materials  = [Ag],       # Introducing metal complex terms
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection, Line3DCollection
from matplotlib.patches import Patch

# =========  =========
# 
# resolution, design_region_x_width, design_region_y_height, design_region_z_height,
# pml_size, Sz_size, Sx, Sy, Sz, Source_distance
# 

# ========= wireframe box =========
def draw_box(ax, xlim, ylim, zlim, color='k', lw=1.5, ls='-'):
    x0, x1 = xlim; y0, y1 = ylim; z0, z1 = zlim
    V = np.array([
        (x0,y0,z0),(x1,y0,z0),(x1,y1,z0),(x0,y1,z0),
        (x0,y0,z1),(x1,y0,z1),(x1,y1,z1),(x0,y1,z1)
    ])
    edges = [(0,1),(1,2),(2,3),(3,0),(4,5),(5,6),(6,7),(7,4),
             (0,4),(1,5),(2,6),(3,7)]
    lines = [[V[i], V[j]] for i,j in edges]
    lc = Line3DCollection(lines, colors=color, linewidths=lw, linestyles=ls)
    ax.add_collection3d(lc)

# ========= // =========
def add_plane(ax, orient='xz', pos=0.0, xlim=(-.5,.5), ylim=(-.5,.5), zlim=(-.3,.3),
              alpha=0.35, color='C0', edge='k', label=None):
    x0,x1 = xlim; y0,y1 = ylim; z0,z1 = zlim
    if orient == 'xy':
        z = pos
        verts = [(x0,y0,z),(x1,y0,z),(x1,y1,z),(x0,y1,z)]
        lab = label or f'xy @ z={pos:.3f} m'
    elif orient == 'xz':
        y = pos
        verts = [(x0,y,z0),(x1,y,z0),(x1,y,z1),(x0,y,z1)]
        lab = label or f'xz @ y={pos:.3f} m'
    elif orient == 'yz':
        x = pos
        verts = [(x,y0,z0),(x,y1,z0),(x,y1,z1),(x,y0,z1)]
        lab = label or f'yz @ x={pos:.3f} m'
    poly = Poly3DCollection([verts], facecolor=color, edgecolor=edge, alpha=alpha)
    ax.add_collection3d(poly)
    return Patch(facecolor=color, edgecolor=edge, label=lab)

# ========= PML slab  =========
def add_block(ax, center, size, alpha=0.35, color='C2', edge='k', label='block'):
    cx,cy,cz = center; sx,sy,sz = size
    x0,x1 = cx - sx/2, cx + sx/2
    y0,y1 = cy - sy/2, cy + sy/2
    z0,z1 = cz - sz/2, cz + sz/2
    faces = [
        [(x0,y0,z0),(x1,y0,z0),(x1,y1,z0),(x0,y1,z0)],  # bottom
        [(x0,y0,z1),(x1,y0,z1),(x1,y1,z1),(x0,y1,z1)],  # top
        [(x0,y0,z0),(x1,y0,z0),(x1,y0,z1),(x0,y0,z1)],  # y0
        [(x0,y1,z0),(x1,y1,z0),(x1,y1,z1),(x0,y1,z1)],  # y1
        [(x0,y0,z0),(x0,y1,z0),(x0,y1,z1),(x0,y0,z1)],  # x0
        [(x1,y0,z0),(x1,y1,z0),(x1,y1,z1),(x1,y0,z1)],  # x1
    ]
    poly = Poly3DCollection(faces, facecolor=color, edgecolor=edge, alpha=alpha)
    ax.add_collection3d(poly)
    return Patch(facecolor=color, edgecolor=edge, label=label)

# =========  3D  =========
def plot_meep_setup_with_slices(
    xy_slices=None, xz_slices=None, yz_slices=None,
    show_design_region=True, show_source=True, title='Meep 3D setup'
):
    fig = plt.figure(figsize=(9,7))
    ax  = fig.add_subplot(111, projection='3d')

    #  cell z  PML
    xlim = (-Sx/2, Sx/2)
    ylim = (-Sy/2, Sy/2)
    zlim = (-Sz/2, Sz/2)
    draw_box(ax, xlim, ylim, zlim, color='gray', lw=1.2, ls='--')

    #  z  PMLx,y  PML 
    z_in = (zlim[0] + pml_size, zlim[1] - pml_size)  #  z  PML
    x_in = xlim
    y_in = ylim
    draw_box(ax, x_in, y_in, z_in, color='k', lw=1.6, ls='-')

    handles = []

    # PML slabs
    handles.append(add_block(ax, (0,0, zlim[0]+pml_size/2), (Sx,Sy,pml_size),
                             alpha=0.12, color='C7', label='PML (z)'))
    handles.append(add_block(ax, (0,0, zlim[1]-pml_size/2), (Sx,Sy,pml_size),
                             alpha=0.12, color='C7', label='PML (+z)'))

    #  MaterialGrid block
    if show_design_region:
        handles.append(add_block(
            ax,
            center=(0,0,0),
            size=(design_region_x_width, design_region_y_height, design_region_z_height),
            alpha=0.4, color='blue', label='Design region'
        ))

    # Ey z=Source_distance
    if show_source:
        handles.append(add_plane(ax, 'xy', Source_distance, xlim, ylim, zlim,alpha=0.7,
                                 color='C3', label=f'Source at z={Source_distance:.3f}'))

    # 
   #if xz_slices:
   #    for y0 in np.atleast_1d(xz_slices):
   #        handles.append(add_plane(ax, 'xz', y0, x_in, y_in, z_in, color='C0',
   #                                 label=f'xz @ y={y0:.3f}'))
   #if xy_slices:
   #    for z0 in np.atleast_1d(xy_slices):
   #        handles.append(add_plane(ax, 'xy', z0, x_in, y_in, z_in, color='C1',
   #                                 label=f'xy @ z={z0:.3f}'))
   #if yz_slices:
   #    for x0 in np.atleast_1d(yz_slices):
   #        handles.append(add_plane(ax, 'yz', x0, x_in, y_in, z_in, color='C4',
   #                                  label=f'yz @ x={x0:.3f}'))

    # 
    ax.set_box_aspect((Sx, Sy, Sz))  # 
    ax.set_xlim(*xlim); ax.set_ylim(*ylim); ax.set_zlim(*zlim)
    ax.set_xlabel('x (m)'); ax.set_ylabel('y (m)'); ax.set_zlabel('z (m)', labelpad=12)
    ax.set_title(title, fontsize=20, pad=30, loc='center')
    if handles:
        #  label  legend 
        uniq = {h.get_label(): h for h in handles}
        ax.legend(handles=list(uniq.values()), loc='upper left', bbox_to_anchor=(0.8,1))
    plt.tight_layout()
    plt.show()

# ===== xz@ y=0 Et(x,z) xy@ z=0 =====
plot_meep_setup_with_slices(
    xz_slices=[0.0],     #  xz  y=0
    xy_slices=[0.0],     #  xy  z=0
    yz_slices=None,      #  yz 
    show_design_region=True,
    show_source=True,
    title='Simulation cell '
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection, Line3DCollection
from matplotlib.patches import Patch

# =========  =========
# 
# resolution, design_region_x_width, design_region_y_height, design_region_z_height,
# pml_size, Sz_size, Sx, Sy, Sz, Source_distance
# 

# ========= wireframe box =========
def draw_box(ax, xlim, ylim, zlim, color='k', lw=1.5, ls='-'):
    x0, x1 = xlim; y0, y1 = ylim; z0, z1 = zlim
    V = np.array([
        (x0,y0,z0),(x1,y0,z0),(x1,y1,z0),(x0,y1,z0),
        (x0,y0,z1),(x1,y0,z1),(x1,y1,z1),(x0,y1,z1)
    ])
    edges = [(0,1),(1,2),(2,3),(3,0),(4,5),(5,6),(6,7),(7,4),
             (0,4),(1,5),(2,6),(3,7)]
    lines = [[V[i], V[j]] for i,j in edges]
    lc = Line3DCollection(lines, colors=color, linewidths=lw, linestyles=ls)
    ax.add_collection3d(lc)

# ========= // =========
def add_plane(ax, orient='xz', pos=0.0, xlim=(-.5,.5), ylim=(-.5,.5), zlim=(-.3,.3),
              alpha=0.35, color='C0', edge='k', label=None):
    x0,x1 = xlim; y0,y1 = ylim; z0,z1 = zlim
    if orient == 'xy':
        z = pos
        verts = [(x0,y0,z),(x1,y0,z),(x1,y1,z),(x0,y1,z)]
        lab = label or f'xy @ z={pos:.3f} m'
    elif orient == 'xz':
        y = pos
        verts = [(x0,y,z0),(x1,y,z0),(x1,y,z1),(x0,y,z1)]
        lab = label or f'xz @ y={pos:.3f} m'
    elif orient == 'yz':
        x = pos
        verts = [(x,y0,z0),(x,y1,z0),(x,y1,z1),(x,y0,z1)]
        lab = label or f'yz @ x={pos:.3f} m'
    poly = Poly3DCollection([verts], facecolor=color, edgecolor=edge, alpha=alpha)
    ax.add_collection3d(poly)
    return Patch(facecolor=color, edgecolor=edge, label=lab)

# ========= PML slab  =========
def add_block(ax, center, size, alpha=0.35, color='C2', edge='k', label='block'):
    cx,cy,cz = center; sx,sy,sz = size
    x0,x1 = cx - sx/2, cx + sx/2
    y0,y1 = cy - sy/2, cy + sy/2
    z0,z1 = cz - sz/2, cz + sz/2
    faces = [
        [(x0,y0,z0),(x1,y0,z0),(x1,y1,z0),(x0,y1,z0)],  # bottom
        [(x0,y0,z1),(x1,y0,z1),(x1,y1,z1),(x0,y1,z1)],  # top
        [(x0,y0,z0),(x1,y0,z0),(x1,y0,z1),(x0,y0,z1)],  # y0
        [(x0,y1,z0),(x1,y1,z0),(x1,y1,z1),(x0,y1,z1)],  # y1
        [(x0,y0,z0),(x0,y1,z0),(x0,y1,z1),(x0,y0,z1)],  # x0
        [(x1,y0,z0),(x1,y1,z0),(x1,y1,z1),(x1,y0,z1)],  # x1
    ]
    poly = Poly3DCollection(faces, facecolor=color, edgecolor=edge, alpha=alpha)
    ax.add_collection3d(poly)
    return Patch(facecolor=color, edgecolor=edge, label=label)

# =========  3D  =========
def plot_meep_setup_with_slices(
    xy_slices=None, xz_slices=None, yz_slices=None,
    show_design_region=True, show_source=True, title='Meep 3D setup'
):
    fig = plt.figure(figsize=(9,7))
    ax  = fig.add_subplot(111, projection='3d')

    #  cell z  PML
    xlim = (-Sx/2, Sx/2)
    ylim = (-Sy/2, Sy/2)
    zlim = (-Sz/2, Sz/2)
    draw_box(ax, xlim, ylim, zlim, color='gray', lw=1.2, ls='--')

    #  z  PMLx,y  PML 
    z_in = (zlim[0] + pml_size, zlim[1] - pml_size)  #  z  PML
    x_in = xlim
    y_in = ylim
    draw_box(ax, x_in, y_in, z_in, color='k', lw=1.6, ls='-')

    handles = []

    # PML slabs
    handles.append(add_block(ax, (0,0, zlim[0]+pml_size/2), (Sx,Sy,pml_size),
                             alpha=0.12, color='C7', label='PML (z)'))
    handles.append(add_block(ax, (0,0, zlim[1]-pml_size/2), (Sx,Sy,pml_size),
                             alpha=0.12, color='C7', label='PML (+z)'))

    #  MaterialGrid block
    if show_design_region:
        handles.append(add_block(
            ax,
            center=(0,0,0),
            size=(design_region_x_width, design_region_y_height, design_region_z_height),
            alpha=0.4, color='blue', label='Design region'
        ))

    # Ey z=Source_distance
    if show_source:
        handles.append(add_plane(ax, 'xy', Source_distance, xlim, ylim, zlim,alpha=0.7,
                                 color='C3', label=f'Source at z={Source_distance:.3f}'))

    # 
   #if xz_slices:
   #    for y0 in np.atleast_1d(xz_slices):
   #        handles.append(add_plane(ax, 'xz', y0, x_in, y_in, z_in, color='C0',
   #                                 label=f'xz @ y={y0:.3f}'))
   #if xy_slices:
   #    for z0 in np.atleast_1d(xy_slices):
   #        handles.append(add_plane(ax, 'xy', z0, x_in, y_in, z_in, color='C1',
   #                                 label=f'xy @ z={z0:.3f}'))
   #if yz_slices:
   #    for x0 in np.atleast_1d(yz_slices):
   #        handles.append(add_plane(ax, 'yz', x0, x_in, y_in, z_in, color='C4',
   #                                  label=f'yz @ x={x0:.3f}'))

    # 
    ax.set_box_aspect((Sx, Sy, Sz))  # 
    ax.set_xlim(*xlim); ax.set_ylim(*ylim); ax.set_zlim(*zlim)
    ax.set_xlabel('x (m)'); ax.set_ylabel('y (m)'); ax.set_zlabel('z (m)', labelpad=12)
    ax.set_title(title, fontsize=20, pad=30, loc='center')
    if handles:
        #  label  legend 
        uniq = {h.get_label(): h for h in handles}
        #ax.legend(handles=list(uniq.values()), loc='upper left', bbox_to_anchor=(0.8,1))
    plt.tight_layout()
    plt.show()

# ===== xz@ y=0 Et(x,z) xy@ z=0 =====
plot_meep_setup_with_slices(
    xz_slices=[0.0],     #  xz  y=0
    xy_slices=[0.0],     #  xy  z=0
    yz_slices=None,      #  yz 
    show_design_region=True,
    show_source=True,
    title='Simulation cell '
)
```
