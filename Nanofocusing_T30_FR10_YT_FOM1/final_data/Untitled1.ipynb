```python
import matplotlib.pyplot as plt

# 建立圖表
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

# 基本格點範圍
nx, ny = 4, 4

# === Yee grid = True ===
axes[0].set_title("Yee grid = True", fontsize=20, pad=25)

# 畫出網格
for i in range(nx+1):
    axes[0].plot([i, i], [0, ny], color="gray", lw=1)
for j in range(ny+1):
    axes[0].plot([0, nx], [j, j], color="gray", lw=1)

# 在不同位置標示 Ex, Ey, Ez
for i in range(nx):
    for j in range(ny+1):
        axes[0].text(i+0.5, j, "Ex", color="red", ha="center", va="center", fontsize=22, fontweight="bold")
for i in range(nx+1):
    for j in range(ny):
        axes[0].text(i, j+0.5, "Ey", color="blue", ha="center", va="center", fontsize=22, fontweight="bold")
for i in range(nx):
    for j in range(ny):
        axes[0].text(i+0.5, j+0.5, "Ez", color="green", ha="center", va="center", fontsize=22, fontweight="bold", alpha=0.7)

axes[0].set_xlim(0, nx)
axes[0].set_ylim(0, ny)
axes[0].set_aspect("equal")
axes[0].axis("off")

# === Yee grid = False ===
axes[1].set_title("Yee grid = False", fontsize=20, pad=25)

# 畫出網格
for i in range(nx):
    for j in range(ny):
        rect = plt.Rectangle((i, j), 1, 1, fill=None, edgecolor="gray", lw=1)
        axes[1].add_patch(rect)

# 在 cell 中心放 Ex, Ey, Ez 疊加
for i in range(nx):
    for j in range(ny):
        axes[1].text(i+0.5, j+0.5, "Ex\nEy\nEz", color="black", ha="center", va="center", fontsize=22, fontweight="bold")

axes[1].set_xlim(0, nx)
axes[1].set_ylim(0, ny)
axes[1].set_aspect("equal")
axes[1].axis("off")

plt.tight_layout()
plt.show()
```
