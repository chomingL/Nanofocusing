```python
import numpy as np
import matplotlib.pyplot as plt

shape = (2, 3, 3)
nx, ny, nz = shape
points = np.array([(i, j, k) for i in range(nx) for j in range(ny) for k in range(nz)])

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:,0], points[:,1], points[:,2], s=60)

# 標註索引 (i,j,k)
for (i, j, k) in points:
    ax.text(i, j, k, f"({i},{j},{k})", fontsize=8, ha='center', va='bottom')

# 隱藏刻度
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

ax.set_xlabel("axis=0 (i)")
ax.set_ylabel("axis=1 (j)")
ax.set_zlabel("axis=2 (k)")
ax.set_title(f"Index points for shape={shape}")
ax.set_box_aspect((nx, ny, nz))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

shape = (3,2,3)
nx, ny, nz = shape
points = np.array([(i, j, k) for i in range(nx) for j in range(ny) for k in range(nz)])

fig = plt.figure(figsize=(6,5.5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:,0], points[:,1], points[:,2], marker='x', s=60)

# 軸標籤
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Case: shape=(3,2,3)", pad=12)

# 軸範圍與刻度（只放在點的位置）
ax.set_xlim(-0.5, nx-0.5)
ax.set_ylim(-0.5, ny-0.5)
ax.set_zlim(-0.5, nz-0.5)
ax.set_xticks(range(nx))
ax.set_yticks(range(ny))
ax.set_zticks(range(nz))

ax.set_box_aspect((nx, ny, nz))
ax.view_init(elev=22, azim=-60)
plt.tight_layout()
plt.show()
```
